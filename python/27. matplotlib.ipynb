```python
import matplotlib.pyplot as plt              # Core plotting interface
import numpy as np                           # For generating sample data
from matplotlib import patches               # For advanced shapes
import matplotlib.dates as mdates            # For date formatting
from datetime import datetime, timedelta     # For date/time data
```

```python
x = np.linspace(0, 10, 100)                  # 100 points from 0 to 10
y = np.sin(x)                                # Sine wave
```

```python
plt.figure(figsize=(6, 4))                  # Set figure size (width, height)
plt.plot(x, y, 
         color='blue',                       # Line color
         linewidth=1,                        # Line thickness
         linestyle='-',                      # Style: '-', '--', '-.', ':'
         marker='o',                         # Marker style: 'o', 's', '^', 'D', '*'
         markevery=10,                       # Show marker every 10 points
         markersize=10,                       # Size of markers
         markerfacecolor='red',              # Marker fill color
         markeredgecolor='black',            # Marker border color
         alpha=0.8,                          # Transparency (0-1)
         label='Sine Wave')                  # Legend label

plt.title('Line Plot - Sine Wave', fontsize=16, fontweight='bold')
plt.xlabel('X Axis (Time/Sequence)', fontsize=12)
plt.ylabel('Y Axis (Values)', fontsize=12)
plt.legend(loc='upper left')                # Legend position: 'upper/lower left/right'
plt.grid(True, alpha=0.3, linestyle='--')    # Grid with transparency
plt.tight_layout()                           # Prevent label cutoff
plt.show()
```

```python
categories = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
values = [23, 45, 56, 78, 32]
colors_bar = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values,
               color=colors_bar,             # Custom colors for each bar
               width=0.6,                    # Bar width (0-1)
               edgecolor='black',            # Border color
               linewidth=1.5,                # Border thickness
               alpha=0.8)                    # Transparency

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}',                # Format: show integer value
             ha='center', va='bottom',       # Horizontal/vertical alignment
             fontsize=10, fontweight='bold')

plt.title('Bar Chart - Product Sales Comparison', fontsize=16, fontweight='bold')
plt.xlabel('Products', fontsize=12)
plt.ylabel('Sales (Units)', fontsize=12)
plt.ylim(0, max(values) * 1.1)               # Set Y-axis limits with padding
plt.grid(axis='y', alpha=0.3)                # Grid only on Y-axis
plt.tight_layout()
plt.show()
```

```python
categories_h = ['Category A', 'Category B', 'Category C', 'Category D']
values_h = [15, 30, 45, 22]

plt.figure(figsize=(10, 6))
bars_h = plt.barh(categories_h, values_h,    # barh = horizontal bar
                  color='skyblue',
                  edgecolor='navy',
                  linewidth=2,
                  alpha=0.7)

# Add value labels at the end of bars
for i, (bar, val) in enumerate(zip(bars_h, values_h)):
    plt.text(val + 1, i, f'{val}',
             va='center', fontsize=10, fontweight='bold')

plt.title('Horizontal Bar Chart', fontsize=16, fontweight='bold')
plt.xlabel('Values', fontsize=12)
plt.ylabel('Categories', fontsize=12)
plt.xlim(0, max(values_h) * 1.2)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
data = np.random.randn(1000) * 10 + 50       # Normal distribution: mean=50, std=10

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data,
                             bins=30,        # Number of bins (intervals)
                             color='green',
                             edgecolor='black',
                             linewidth=1,
                             alpha=0.7,
                             density=False,  # False=counts, True=probability density
                             cumulative=False) # True for cumulative histogram
```

```python

plt.figure(figsize=(10, 6))
n, bins, patches = plt.hist(data,
                             bins=30,        # Number of bins (intervals)
                             color='green',
                             edgecolor='black',
                             linewidth=1,
                             alpha=0.7,
                             density=False,  # False=counts, True=probability density
                             cumulative=False) # True for cumulative histogram

# Color bars by height
for i, patch in enumerate(patches):
    patch.set_facecolor(plt.cm.viridis(n[i]/max(n)))

plt.axvline(np.mean(data), color='red', linestyle='--', 
            linewidth=2, label=f'Mean = {np.mean(data):.2f}')
plt.axvline(np.median(data), color='orange', linestyle='--',
            linewidth=2, label=f'Median = {np.median(data):.2f}')

plt.title('Histogram - Data Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
sizes = [30, 25, 20, 15, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
colors_pie = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
explode = (0.1, 0, 0, 0, 0)                  # "Explode" first slice

plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(sizes,
                                    labels=labels,
                                    colors=colors_pie,
                                    autopct='%1.1f%%',   # Show percentage
                                    startangle=90,       # Rotate start angle
                                    explode=explode,     # Pull out slices
                                    shadow=True,         # Add shadow effect
                                    textprops={'fontsize': 11})

# Make percentage text bold
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.title('Pie Chart - Market Share Distribution', fontsize=16, fontweight='bold')
plt.axis('equal')                            # Equal aspect ratio = circular pie
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(sizes,
                                    labels=labels,
                                    colors=colors_pie,
                                    autopct='%1.1f%%',
                                    startangle=90,
                                    pctdistance=0.85,    # Position of percentage text
                                    wedgeprops={'width': 0.4, 'edgecolor': 'white'}) # Creates donut

# Add center circle for donut effect
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Donut Chart - Proportional Data', fontsize=16, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
np.random.seed(10)
data_box = [np.random.normal(0, std, 100) for std in range(1, 5)]

plt.figure(figsize=(6, 4))
bp = plt.boxplot(data_box,
                 labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'],
                 notch=True,                 # Notched box shows confidence interval
                 patch_artist=True,          # Fill boxes with color
                 showmeans=True,             # Show mean as well as median
                 meanline=True,              # Show mean as line
                 widths=0.5)                 # Box width

# Customize box colors
colors_box = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
for patch, color in zip(bp['boxes'], colors_box):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

plt.title('Box Plot - Distribution Comparison', fontsize=16, fontweight='bold')
plt.xlabel('Groups', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
bp = plt.boxplot(data_box,
                 labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'],
                 notch=True,                 # Notched box shows confidence interval
                 patch_artist=True,          # Fill boxes with color
                 showmeans=True,             # Show mean as well as median
                 meanline=True,              # Show mean as line
                 widths=0.5)                 # Box width
plt.title('Box Plot - Distribution Comparison', fontsize=16, fontweight='bold')
plt.xlabel('Groups', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
parts = plt.violinplot(data_box,
                       positions=[1, 2, 3, 4],
                       showmeans=True,       # Show mean marker
                       showmedians=True,     # Show median line
                       widths=0.7)           # Violin width

# Customize violin colors
for pc in parts['bodies']:
    pc.set_facecolor('skyblue')
    pc.set_alpha(0.7)
    pc.set_edgecolor('navy')

plt.title('Violin Plot - Distribution with Density', fontsize=16, fontweight='bold')
plt.xlabel('Groups', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks([1, 2, 3, 4], ['Group 1', 'Group 2', 'Group 3', 'Group 4'])
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
categories_stack = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [10, 15, 13, 17]
product_b = [12, 18, 14, 16]
product_c = [8, 12, 15, 19]

plt.figure(figsize=(10, 6))
bar_width = 0.6
plt.bar(categories_stack, product_a, bar_width, 
        label='Product A', color='#FF6B6B')
plt.bar(categories_stack, product_b, bar_width, 
        bottom=product_a, label='Product B', color='#4ECDC4')
plt.bar(categories_stack, product_c, bar_width,
        bottom=np.array(product_a) + np.array(product_b),
        label='Product C', color='#45B7D1')

plt.title('Stacked Bar Chart - Quarterly Sales', fontsize=16, fontweight='bold')
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
categories_group = ['A', 'B', 'C', 'D']
group1 = [20, 35, 30, 35]
group2 = [25, 32, 34, 20]
group3 = [18, 30, 25, 30]

x_pos = np.arange(len(categories_group))
bar_width = 0.25

plt.figure(figsize=(10, 6))
plt.bar(x_pos - bar_width, group1, bar_width, 
        label='2022', color='#FF6B6B', edgecolor='black')
plt.bar(x_pos, group2, bar_width,
        label='2023', color='#4ECDC4', edgecolor='black')
plt.bar(x_pos + bar_width, group3, bar_width,
        label='2024', color='#45B7D1', edgecolor='black')

plt.title('Grouped Bar Chart - Year Comparison', fontsize=16, fontweight='bold')
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(x_pos, categories_group)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
x = np.arange(0, 10, 1)
y = np.random.randint(10, 50, 10)
y_error = np.random.randint(2, 8, 10)        # Error margins
x_error = np.random.randint(0, 3, 10)        # X-axis errors (optional)

plt.figure(figsize=(10, 6))
plt.errorbar(x, y,
             yerr=y_error,                   # Y-axis error bars
             #xerr=x_error,                   # X-axis error bars
             fmt='o',                        # Format: 'o', 's', '^'
             ecolor='red',                   # Error bar color
             elinewidth=2,                   # Error bar line width
             capsize=5,                      # Cap size at ends
             capthick=2,                     # Cap thickness
             alpha=0.7,
             label='Data with errors')

plt.title('Error Bar Plot - Data with Uncertainty', fontsize=16, fontweight='bold')
plt.xlabel('X Axis', fontsize=12)
plt.ylabel('Y Axis', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
data_heat = np.random.rand(10, 10)           # 10x10 random matrix

plt.figure(figsize=(10, 8))
im = plt.imshow(data_heat,
                cmap='YlOrRd',               # Color map: 'hot', 'cool', 'RdYlGn'
                aspect='auto',               # Aspect ratio
                interpolation='nearest')     # Interpolation: 'nearest', 'bilinear'

plt.colorbar(im, label='Values')

# Add value labels in cells
for i in range(10):
    for j in range(10):
        text = plt.text(j, i, f'{data_heat[i, j]:.2f}',
                       ha="center", va="center",
                       color="black", fontsize=8)

plt.title('Heatmap - Matrix Visualization', fontsize=16, fontweight='bold')
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Rows', fontsize=12)
plt.tight_layout()
plt.show()
```
