```python
import plotly.graph_objects as go          # Core plotting objects
import plotly.express as px                # High-level interface (easier)
import pandas as pd                        # Data manipulation
import numpy as np                         # Numerical operations
from plotly.subplots import make_subplots  # For multiple subplots
import plotly.io as pio                    # For saving plots
```

```python
# Set default template for all plots
pio.templates.default = "plotly"           # Templates: 'plotly', 'plotly_white', 'plotly_dark', 'ggplot2', 'seaborn', 'simple_white'

# Set random seed for reproducibility
np.random.seed(42)
```

```python
# ---- Large Sales Dataset (1500 rows) ----------------------------------------
sales_data = pd.DataFrame({
    'Date': pd.date_range('2021-01-01', periods=1500, freq='D'),
    'Region': np.random.choice(['North America', 'Europe', 'Asia', 'South America', 'Africa'], 1500),
    'Product': np.random.choice(['Laptop', 'Phone', 'Tablet', 'Watch', 'Headphones'], 1500),
    'Category': np.random.choice(['Electronics', 'Accessories', 'Wearables'], 1500),
    'Sales': np.random.randint(1000, 10000, 1500) + np.random.randn(1500) * 500,
    'Profit': np.random.randint(100, 2000, 1500) + np.random.randn(1500) * 100,
    'Quantity': np.random.randint(1, 100, 1500),
    'Discount': np.random.choice([0, 5, 10, 15, 20, 25], 1500),
    'Customer_Age': np.random.randint(18, 75, 1500),
    'Rating': np.random.uniform(1, 5, 1500),
    'Price': np.random.randint(50, 2000, 1500)
})
sales_data['Revenue'] = sales_data['Sales'] * sales_data['Price']

# ---- Employee Dataset (800 rows) --------------------------------------------
employee_data = pd.DataFrame({
    'Employee_ID': range(1, 801),
    'Department': np.random.choice(['IT', 'HR', 'Sales', 'Marketing', 'Finance', 'Operations'], 800),
    'Salary': np.random.randint(30000, 200000, 800),
    'Experience': np.random.randint(0, 30, 800),
    'Age': np.random.randint(22, 70, 800),
    'Gender': np.random.choice(['Male', 'Female', 'Other'], 800),
    'Education': np.random.choice(['Bachelor', 'Master', 'PhD', 'High School'], 800),
    'Performance_Score': np.random.uniform(1, 10, 800),
    'Satisfaction': np.random.randint(1, 11, 800),
    'Bonus': np.random.randint(0, 30000, 800),
    'Projects_Completed': np.random.randint(0, 50, 800)
})

# ---- Stock Market Dataset (3 years) -----------------------------------------
dates_stock = pd.date_range('2021-01-01', '2023-12-31', freq='D')
stock_data = pd.DataFrame({
    'Date': dates_stock,
    'AAPL': 150 + np.cumsum(np.random.randn(len(dates_stock))) * 2,
    'GOOGL': 2800 + np.cumsum(np.random.randn(len(dates_stock))) * 20,
    'MSFT': 300 + np.cumsum(np.random.randn(len(dates_stock))) * 3,
    'TSLA': 800 + np.cumsum(np.random.randn(len(dates_stock))) * 15,
    'AMZN': 3400 + np.cumsum(np.random.randn(len(dates_stock))) * 25
})
stock_data['Volume_AAPL'] = np.random.randint(50000000, 150000000, len(dates_stock))

# ---- Geographic Dataset (World Cities) --------------------------------------
cities_data = pd.DataFrame({
    'City': ['New York', 'London', 'Tokyo', 'Paris', 'Sydney', 'Mumbai', 'Dubai', 
             'Singapore', 'Toronto', 'Berlin', 'Beijing', 'Sao Paulo', 'Moscow',
             'Los Angeles', 'Hong Kong', 'Seoul', 'Madrid', 'Rome', 'Amsterdam', 'Bangkok'],
    'Latitude': [40.71, 51.51, 35.68, 48.86, -33.87, 19.08, 25.20, 
                 1.35, 43.65, 52.52, 39.90, -23.55, 55.75,
                 34.05, 22.32, 37.57, 40.42, 41.90, 52.37, 13.75],
    'Longitude': [-74.01, -0.13, 139.65, 2.35, 151.21, 72.88, 55.27,
                  103.82, -79.38, 13.41, 116.41, -46.63, 37.62,
                  -118.24, 114.17, 126.98, -3.70, 12.50, 4.90, 100.52],
    'Population': [8800000, 9000000, 14000000, 2200000, 5300000, 20400000, 3400000,
                   5700000, 2900000, 3700000, 21500000, 12300000, 12500000,
                   4000000, 7500000, 9700000, 3200000, 2900000, 870000, 10500000],
    'GDP': [1700, 650, 1600, 720, 480, 310, 400, 
            340, 380, 460, 610, 430, 520,
            1000, 450, 780, 230, 150, 110, 200]
})

# ---- 3D Dataset for Surface/Mesh Plots --------------------------------------
x_3d = np.linspace(-5, 5, 50)
y_3d = np.linspace(-5, 5, 50)
X_3d, Y_3d = np.meshgrid(x_3d, y_3d)
Z_3d = np.sin(np.sqrt(X_3d**2 + Y_3d**2))

# ---- Network/Hierarchy Dataset ----------------------------------------------
network_data = pd.DataFrame({
    'Source': ['CEO', 'CEO', 'CEO', 'CTO', 'CTO', 'CFO', 'CFO', 'VP Sales', 'VP Sales'],
    'Target': ['CTO', 'CFO', 'VP Sales', 'Dev Team', 'QA Team', 'Accounting', 'Finance', 'Sales Team A', 'Sales Team B'],
    'Value': [50, 40, 30, 25, 20, 20, 20, 15, 15]
})

print("✓ Demo datasets created successfully!")
print(f"✓ Sales Data: {len(sales_data)} rows")
print(f"✓ Employee Data: {len(employee_data)} rows")
print(f"✓ Stock Data: {len(stock_data)} rows")
print(f"✓ Cities Data: {len(cities_data)} rows\n")
```

```python
fig = px.scatter(sales_data,
                 x='Sales',                  # X-axis variable
                 y='Profit',                 # Y-axis variable
                 color='Region',             # Color by category
                 #size='Quantity',            # Bubble size
                 hover_data=['Product', 'Rating', 'Discount'],  # Extra info on hover
                 title='Interactive Scatter Plot - Sales vs Profit Analysis',
                 labels={'Sales': 'Sales Amount ($)', 'Profit': 'Profit ($)'},  # Custom labels
                 color_discrete_sequence=px.colors.qualitative.Set2,  # Color palette
                 #size_max=30,                # Maximum bubble size
                 opacity=0.7,                # Transparency
                 template='plotly_white')    # Template style

# Customize layout
fig.update_layout(
    font=dict(size=12),
    title_font_size=16,
    hovermode='closest',                     # Hover mode: 'closest', 'x', 'y', 'x unified'
    height=600,
    width=1000
)

# Add custom hover template
fig.update_traces(
    hovertemplate='<b>Sales:</b> $%{x:.2f}<br><b>Profit:</b> $%{y:.2f}<br><b>Quantity:</b> %{marker.size}<extra></extra>'
)

fig.show()
```

```python
fig = px.line(stock_data,
              x='Date',
              y=['AAPL', 'GOOGL', 'MSFT', 'TSLA', 'AMZN'],  # Multiple series
              title='Interactive Line Chart - Stock Price Trends',
              labels={'value': 'Stock Price ($)', 'variable': 'Company'},
              template='plotly_dark')

# Customize layout with range slider and buttons
fig.update_layout(
    hovermode='x unified',                   # Show all values at same x-position
    height=600,
    xaxis=dict(
        rangeselector=dict(                  # Add date range selector buttons
            buttons=list([
                dict(count=1, label="1M", step="month", stepmode="backward"),
                dict(count=6, label="6M", step="month", stepmode="backward"),
                dict(count=1, label="1Y", step="year", stepmode="backward"),
                dict(step="all", label="All")
            ])
        ),
        rangeslider=dict(visible=True),      # Add range slider at bottom
        type="date"
    )
)

fig.show()
```

```python
bar_data = sales_data.groupby('Region')['Revenue'].sum().reset_index()

fig = px.bar(bar_data,
             x='Region',
             y='Revenue',
             color='Region',
             title='Interactive Bar Chart - Total Revenue by Region',
             labels={'Revenue': 'Total Revenue ($)'},
             text='Revenue',                 # Show values on bars
             template='seaborn',
             color_discrete_sequence=px.colors.qualitative.Pastel)

# Customize bars and layout
fig.update_traces(
    texttemplate='$%{text:,.0f}',           # Format text on bars
    textposition='outside',                  # Position: 'inside', 'outside', 'auto'
    marker_line_color='black',              # Bar border color
    marker_line_width=1.5
)

fig.update_layout(
    height=600,
    showlegend=False,                        # Hide legend (redundant with x-axis)
    yaxis_title='Revenue ($)',
    xaxis_title='Region'
)

fig.show()
```

```python
plt.figure(figsize=(6,10))
fig = px.histogram(employee_data,
                   x='Salary',
                   nbins=40,                 # Number of bins
                   color='Department',       # Stack by category
                   title='Interactive Histogram - Salary Distribution by Department',
                   labels={'Salary': 'Annual Salary ($)'},
                   marginal='box',           # Add marginal plot: 'box', 'violin', 'rug'
                   opacity=0.7,
                   barmode='overlay',        # Bar mode: 'overlay', 'stack', 'group'
                   template='plotly_white')

fig.show()
```

```python
fig = px.box(sales_data,
             x='Region',
             y='Sales',
             #color='Product',
             title='Interactive Box Plot - Sales Distribution by Region and Product',
             notched=True,                   # Show confidence interval notch
             points='outliers',              # Show points: 'all', 'outliers', 'suspectedoutliers', False
             hover_data=['Customer_Age', 'Rating'],
             template='plotly')

fig.update_layout(height=600, boxmode='group')  # Box mode: 'group', 'overlay'
fig.show()
```

```python
pie_data = sales_data.groupby('Product')['Revenue'].sum().reset_index()

fig = px.pie(pie_data,
             values='Revenue',
             names='Product',
             title='Interactive Pie Chart - Revenue Distribution by Product',
             hole=0.6,                       # Creates donut chart (0-1, 0=full pie)
             color_discrete_sequence=px.colors.sequential.RdBu,
             template='seaborn')

# Customize pie chart
fig.update_traces(
    textposition='inside',
    textinfo='percent+label',                # Options: 'label', 'percent', 'value', 'text'
    hovertemplate='<b>%{label}</b><br>Revenue: $%{value:,.0f}<br>Percentage: %{percent}<extra></extra>',
    marker=dict(line=dict(color='white', width=2))  # Add white border
)

fig.update_layout(height=600)
fig.show()
```

```python
fig = px.sunburst(sales_data,
                  path=['Region', 'Product', 'Category'],  # Hierarchy levels
                  values='Revenue',
                  color='Profit',
                  color_continuous_scale='RdYlGn',
                  title='Interactive Sunburst - Hierarchical Revenue Analysis',
                  hover_data=['Quantity'],
                  template='plotly_white')

fig.update_layout(height=700)
fig.show()
```

```python
fig = px.treemap(sales_data,
                 path=['Region', 'Product'],  # Hierarchy
                 values='Revenue',
                 color='Profit',
                 color_continuous_scale='Viridis',
                 title='Interactive Treemap - Revenue and Profit Analysis',
                 hover_data=['Quantity', 'Rating'],
                 template='plotly')

fig.update_layout(height=400)
fig.update_traces(
    textposition='middle center',
    textfont_size=12,
    marker=dict(line=dict(width=2, color='white'))
)

fig.show()
```
