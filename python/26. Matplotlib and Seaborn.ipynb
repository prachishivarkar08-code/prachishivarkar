```python
import seaborn as sns       # Import seaborn library for advanced statistical plotting
import matplotlib.pyplot as plt  # Import matplotlib for additional plot customization and display
import numpy as np          # Import numpy for numerical array operations and data generation
import pandas as pd         # Import pandas for DataFrame operations (seaborn works seamlessly with DataFrames)
import warnings
warnings.filterwarnings("ignore")
```

```python
tips = sns.load_dataset("tips")         # Load restaurant tips dataset with bill amounts, tips, and customer info
iris = sns.load_dataset("iris")         # Load iris flower dataset - classic for machine learning classification
titanic = sns.load_dataset("titanic")   # Load Titanic passenger dataset with survival and demographic data
flights = sns.load_dataset("flights")   # Load airline passenger data over multiple years and months
```

```python
# Use a scatter plot to show relationship/correlation between two continuous variables.
sns.scatterplot(data=tips,              # Use the tips dataset
                x="total_bill",          # Set x-axis to total bill amount
                y="tip",                 # Set y-axis to tip amount
                hue="sex",               # Color points by customer gender
                style="time",            # Use different markers for lunch vs dinner
                size="size",             # Make point size represent party size
                sizes=(20, 200))         # Set minimum and maximum point sizes
plt.title("Scatter Plot: Total Bill vs Tip (Grouped by Sex/Time, Sized by Group)")  # Add descriptive title
plt.xlabel("Total Bill ($)")            # Label x-axis with units for clarity
plt.ylabel("Tip ($)")                   # Label y-axis with units for clarity
plt.show()                              # Display the completed plot
```

```python
# Line plots are for trends (time series, sequences, or continuous data changes).
plt.figure(figsize=(8, 4))
sns.lineplot(data=flights,              # Use the flights dataset
             x="year",                  # Set x-axis to year (time progression)
             y="passengers")           # Set y-axis to number of passengers
plt.title("Line Plot: Airline Passengers Over Years")  # Add title explaining the trend
plt.xlabel("Year")                      # Label x-axis
plt.ylabel("Number of Passengers")      # Label y-axis
plt.show()                              # Display the plot
```

```python
# Example: Grouped line plot for advanced trend analysis
plt.figure(figsize=(8, 4))
sns.lineplot(data=flights,              # Use the flights dataset again
             x="year",                  # Set x-axis to year
             y="passengers",           # Set y-axis to passengers
             hue="month")               # Create separate lines for each month
plt.title("Line Plot: Passengers per Month by Year")  # Title for multi-line plot
plt.xlabel("Year")                      # Label x-axis
plt.ylabel("Number of Passengers")      # Label y-axis
plt.legend(title="Month")               # Show legend with month labels
plt.show()                              # Display the multi-line plot
```

```python
# Bar plots compare numeric values across categories (mean/aggregate per category).
plt.figure(figsize=(8, 4))
sns.barplot(data=titanic,               # Use the titanic dataset
            x="class",                  # Set x-axis to passenger class categories
            y="fare",                   # Set y-axis to fare amounts (will show averages)
            hue="sex",                  # Group bars by gender
            ci=None,                    # Remove confidence intervals for cleaner look
            palette="muted" )  # Use muted color palette
                      
plt.title("Bar Plot: Average Fare by Class and Gender (Titanic)")  # Descriptive title
plt.xlabel("Class")                     # Label x-axis
plt.ylabel("Average Fare")              # Label y-axis (clarifies it shows averages)
plt.show()                              # Display the bar plot
```

```python
# Histograms show distribution of a numeric variable (frequency of value bins).
plt.figure(figsize=(8, 5))
sns.histplot(data=tips,                 # Use the tips dataset
             x="total_bill",            # Set variable to analyze
             bins=10,                   # Divide data into 20 bins for histogram
             kde=True,                  # Add kernel density estimate curve
             color="teal")              # Set histogram color to teal
plt.title("Histogram + KDE: Total Bill Distribution")  # Title explaining both histogram and KDE
plt.xlabel("Total Bill ($)")            # Label x-axis with units
plt.ylabel("Frequency")                 # Label y-axis to show frequency count
plt.show()                              # Display the histogram
```

```python
# Box plots visualize distribution, median, outliers, and compare groups/categories.
plt.figure(figsize=(8, 5))
sns.boxplot(data=tips,                  # Use the tips dataset
            x="day",                    # Set x-axis to day of week categories
            y="total_bill",             # Set y-axis to total bill amounts
            hue="time",                 # Group boxes by lunch/dinner time
            palette="Set3")             # Use Set3 color palette
plt.title("Box Plot: Total Bill by Day and Time")  # Descriptive title
plt.xlabel("Day of Week")               # Label x-axis
plt.ylabel("Total Bill ($)")            # Label y-axis with units
plt.show()                              # Display the box plot
```

```python
# Violin plots show distribution, density, and median like a smoothed boxplot.
plt.figure(figsize=(8, 5))
sns.violinplot(x="species",             # Set x-axis to iris species categories
               y="petal_length",        # Set y-axis to petal length measurements
               data=iris,               # Use the iris dataset
               palette="pastel")        # Use pastel color palette
plt.title("Violin Plot: Petal Length by Species (Iris)")  # Descriptive title
plt.xlabel("Species")                   # Label x-axis
plt.ylabel("Petal Length (cm)")         # Label y-axis with units
plt.show()                              # Display the violin plot
```

```python
# Heatmaps visualize matrix-style data (correlation/aggregation tables).
plt.figure(figsize=(8, 5))
corr = tips.corr(numeric_only=True)     # Calculate correlation matrix of numeric columns only
sns.heatmap(corr,                       # Create heatmap from correlation matrix
            annot=True,                 # Show correlation values in each cell
            cmap="YlGnBu",              # Use Yellow-Green-Blue color map
            fmt=".2f")                  # Format numbers to 2 decimal places
plt.title("Heatmap: Correlation Matrix of Tips Data")  # Descriptive title
plt.show()                              # Display the heatmap
```

```python
# Example: Pivot table for heatmap
flights_pivot = flights.pivot( columns= "month",   # Set rows to months
                      index= "year",    # Set columns to years
                         values=  "passengers")  # Fill cells with passenger counts
```

```python
flights_pivot
```

```python

sns.heatmap(flights_pivot,              # Create heatmap from pivot table
            cmap='Blues',               # Use blues color map
            annot=True,                 # Show passenger numbers in cells
            fmt='d')                    # Format numbers as integers (no decimals)
plt.title("Flight Passengers per Month (annotated)")  # Descriptive title
plt.xlabel("Year")                      # Label x-axis
plt.ylabel("Month")                     # Label y-axis
plt.show()                              # Display the pivot heatmap
```

```python
# Pair plots explore pairwise relationships among numeric variables.
sns.pairplot(iris,                      # Use the iris dataset
             hue="species",             # Color points by species
             diag_kind="kde",           # Use KDE plots on diagonal (instead of histograms)
             palette="husl")            # Use HUSL color palette
plt.suptitle("Pair Plot: Iris Dataset", y=1.02)  # Add main title slightly above the grid
plt.show()                              # Display the pair plot grid
```

```python
# Joint plots combine scatter plots and histograms for two variables at once.
sns.jointplot(data=iris,                # Use the iris dataset
              x='sepal_length',         # Set x-axis variable
              y='sepal_width',          # Set y-axis variable
            hue='species',         # Color points by species
              kind='scatter')           # Use scatter plot in center (other options: 'hex', 'kde')
plt.suptitle("Joint Plot: Sepal Length vs Sepal Width (Iris)", y=1.02)  # Add main title above
plt.show()                              # Display the joint plot
```

```python
#                            When to Use Which Plot:
#  - Scatter Plot: Relationship/correlation (continuous variables, categories via hue/size).
#  - Line Plot: Trends over a sequence (time, continuous index).
#  - Bar Plot: Compare groups or categories (mean/aggregate metric).
#  - Histogram/KDE: Distribution of a numeric variable (frequency, density).
#  - Box/Violin Plot: Distribution + comparison across categorical groups (spread, outliers).
#  - Heatmap: Matrix data patterns (correlation, cross-tab, pivot tables).
#  - Pair Plot: Explore multiple variable relationships (EDA, clustering/classification).
#  - Joint Plot: Deep dive on two variables' relationship + their marginals.
```
