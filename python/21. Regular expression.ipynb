```python
text = "Contact us at support_2@example.com or sales@mydomain.org, sales@mydomain123.co.in"

pattern = r'\b[\w.-]+@[\w.-]+\.\w+\b'

emails = re.findall(pattern, text)

print(emails)

# r'\b\w+@\w+\.\w+'
# OR
# '\\b\\w+@\\w+\\.\\w+'
```

```python
text = "Call me at 9876543210 or 919876543210, +915888787750."

phones = re.findall(r'\b(?:\+91|91)?[6-9]\d{9}\b', text)

print(phones)
```

```python
# Validate if a String is a Strong Password
password = "MY&tr0ngp@ss"

pattern = r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@$#]).{8,}$'

print(bool(re.match(pattern, password)))

#Checks for uppercase, lowercase, digit, special character, and min length.
```

```python
#Find Dates in dd-mm-yyyy Format
text = "The events are on 21-06-2025 and 15-08-2024, 2026-01-12."

dates = re.findall(r'\b\d{2,4}-\d{2}-\d{2,4}\b', text)

print(dates)
```
